Result of inferences====================It is also easy to use inference as a routine in more complex procedure.

```python
r=range(0,100,2)xs=[x/100.0 for x in r]ys=[gumnb.getPosterior(bn,{'MINVOLSET':[0,x/100.0,0.5]},'VENTALV').tolist()         for x in r]p=plot(xs,ys)legend(p,[bn.variableFromName('VENTALV').label(i)           for i in range(len(bn.variableFromName('VENTALV')))],loc=7);title('VENTALV');ylabel('posterior Probability');xlabel('Evidence on MINVOLSET : [0,x,0.5]');
```
