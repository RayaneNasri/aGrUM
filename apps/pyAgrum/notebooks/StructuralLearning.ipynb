```python
%matplotlib inline
from pylab import *
import matplotlib.pyplot as plt
```

```python
import pyAgrum as gum
import gumLib.notebook as gnb
```

# Generating the database from a BN

```python
bn=gum.loadBN("../resources/asia.bif")
gnb.showBN(bn)
print bn.variableFromName("visit_to_Asia?")
```

```python
# generateCSV from a BN : the loglikelihood of the base is returned
import gumLib.bn2csv as bn2csv
bn2csv.generateCSV(bn,"sample_asia.csv",500000)
```

# Learning a BN from the database

```python
learn=gum.BNLearnerFromCSV()
bn2=learn.LocalSearchWithTabu("sample_asia.csv")
gnb.showBN(bn2)
kl=gum.BruteForceKL(bn,bn2)
kl.compute()
```

And a greedy Hill Climbing algorithm (with insert, remove and change arc as atomic operations).

```python
bn2=learn.GreedyHillClimbing("sample_asia.csv")
gnb.showBN(bn2)
kl=gum.BruteForceKL(bn,bn2)
kl.compute()
```

# Impact of the size of the database for the learning

```python
import IPython.display
rows=3
sizes=[200,250,300,400,500,700,1000,2000,5000,
       10000,20000,50000,75000,
       100000,150000,175000,
       200000,300000,500000]
```

```python
res="<table>"
nbr=0
l=[]
for i in sizes:
    n=i+1
    !head -n $n sample_asia.csv > extract_asia.csv
    #!wc extract_asia.csv
    bn2=learn.GreedyHillClimbing("extract_asia.csv")
    !rm -f extract_asia.csv
    
    kl=gum.BruteForceKL(bn,bn2)
    r=kl.compute()
    l.append(r['klPQ'])
    
    if nbr % rows == 0:
        res+="<tr>"
    res+="<td>"+gnb.getBN(bn2,size="3").data+"</td>"
    nbr+=1
    if nbr % rows == 0:
        res+="</tr>"
if nbr % rows!=0:
    res+="</tr>"
res+="</table>"

IPython.display.display(IPython.display.HTML(res))

plot(sizes,l)
print(l[-1])
```
