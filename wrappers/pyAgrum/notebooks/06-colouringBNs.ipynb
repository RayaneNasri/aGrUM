```python
%matplotlib inline
from pylab import *
import matplotlib.pyplot as plt
```

```python
import pyAgrum as gum
import pyAgrum.lib.notebook as gnb

bn=gum.fastBN("a->b->c->d;b->e->d->f;g->c")
gnb.sideBySide(bn,gnb.getInference(bn))
```

```python
def value(n):
    return 0.5 if n in "aeiou" else 0.7
def arcvalue(a):
    return (10-a[0])*a[1]
gnb.showBN(bn,
           vals={n:value(n) for n in bn.names()},
           arcvals={a:arcvalue(a) for a in bn.arcs()})
```

```python
gnb.showInference(bn,
           targets={"a","g","f","b"},
           vals={n:value(n) for n in bn.names()},
           arcvals={a:arcvalue(a) for a in bn.arcs()})
```

```python
gnb.sideBySide(gnb.getBN(bn,
                         vals={n:value(n) for n in bn.names()},
                         arcvals={a:arcvalue(a) for a in bn.arcs()}),
               gnb.getInference(bn,
                                vals={n:value(n) for n in bn.names()},
                                arcvals={a:arcvalue(a) for a in bn.arcs()})
              )
```

```python
import matplotlib.pyplot as plt
mycmap=plt.get_cmap('Pastel1')
gnb.sideBySide(gnb.getBN(bn,
                         vals={n:value(n) for n in bn.names()},
                         arcvals={a:arcvalue(a) for a in bn.arcs()},
                         cmap=mycmap),
               gnb.getInference(bn,
                                vals={n:value(n) for n in bn.names()},
                                arcvals={a:arcvalue(a) for a in bn.arcs()},
                                cmap=mycmap)
              )
```
