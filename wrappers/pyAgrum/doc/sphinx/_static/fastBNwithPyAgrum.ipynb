```python
import pyAgrum as gum
import pyAgrum.lib.notebook as gnb

import pyAgrum.causal as csl
import pyAgrum.causal.notebook as cslnb

from IPython.display import HTML
```

```python
bn=gum.fastBN("A->B;C->B->D",3)
gnb.sideBySide(bn,
               bn.cpt("B"),
               gnb.getJunctionTree(bn),
               gnb.getInference(bn),
               captions=['a BN','a CPT : $P(B|A,C)$','a Junction Tree','an inference'])
```

```python
model=gum.fastID("*c<-*a->b<-c->$d;b->$d",3)

gnb.sideBySide(model,
               model.utility("d"),
               gnb.getInference(model),
               captions=['an Influence Diagram','a utility : $U(d|b,c)$','an inference'])
```

```python
model=gum.fastMN("a--b--c;b--c--d",3)

gnb.sideBySide(model,
               model.factor(['a','b','c']),
               gnb.getInference(model),
               captions=[' Markov Net (as factor graph)','a factor : $\phi(a,b,c)$','an inference'])
```

```python
bn=gum.fastBN("A->B;C->B->D",3)
d = csl.CausalModel(bn, [("lat1", ["A","D"])])
cslnb.showCausalImpact(d,"D","A",values={"A":0})
```

```python
bn=gum.fastBN("A->B;C->B->D")
bn_min=gum.BayesNet(bn)
bn_max=gum.BayesNet(bn)
for n in bn.nodes():
  x=0.4*min(bn.cpt(n).min(),1-bn.cpt(n).max())
  bn_min.cpt(n).translate(-x)
  bn_max.cpt(n).translate(x)
```

```python
cn=gum.CredalNet(bn_min,bn_max)
cn.intervalToCredal()
cn.computeBinaryCPTMinMax()
gnb.sideBySide(cn,
               HTML(f'<pre style="background-color:#F1F1FF;">{cn.__str__()}</pre>'),
               gnb.getInference(cn,gum.CNLoopyPropagation(cn)),
              captions=["a Credal Network","Lower/Upper probabilities","L2U inference with lower/upper posterior"])
```
